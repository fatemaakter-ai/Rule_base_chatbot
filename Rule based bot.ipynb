```python
import re 
import random
```

```python
class RuleBot:
    # Potential negative response
    negative_response=('no', 'nope', 'nah', 'not a chance', 'sorry', 'nay', 'never')
    # Exit conversation keyword
    exit_commands=('quit','pause','exit','goodbye','bye','later')
    #Random starter Question
    random_question =[
        "Could you describe your business or project and its objectives?\n",
        "What specific problems or challenges are you looking to address with deep learning?\n",
        "Do you have any existing data or datasets that we can work with, or would data collection be part of the project?\n",
        "What is your expected timeline for this deep learning project?\n",
        "What are your budget constraints or considerations for this project?\n",
        "Have you previously worked on any AI or deep learning projects, and if so, what were the outcomes?\n",
        "Do you have a preference for any particular deep learning frameworks or technologies?\n",
        "Are there any specific performance metrics or key performance indicators (KPIs) you'd like to achieve through this project?\n",
        "What level of expertise or support do you currently have in-house for AI and deep learning?",
        "Are there any regulatory or compliance considerations we should be aware of?",
        "Are there potential scalability requirements for the deep learning models or solutions we develop?",
        "What is your vision for the long-term use of AI and deep learning in your organization?",
        "What are your data privacy and security concerns, if any?",
        "Are there any preferred communication and collaboration methods for this project?",
        "Do you have a preferred project management approach or methodology?",
        "What is the primary business impact or outcome you hope to achieve with the implementation of deep learning?",
        "Are there any specific industries or applications within your business where you see potential for deep learning solutions?",
        "Can you provide examples of any successful deep learning projects you've seen or heard about that align with your goals?",
        "How would you measure the success of this deep learning project?",
        "What is your preferred method for data acquisition and preprocessing?",
        "Are there any constraints or limitations on the hardware infrastructure you can use for deep learning?",
        "Do you have any preference for on-premises or cloud-based deep learning solutions?",
        "What are your thoughts on model interpretability and explainability for the solutions we develop?",
        "Are there any specific stakeholders or departments within your organization that will be closely involved in this project?",
        "Do you anticipate any challenges related to obtaining labeled or annotated data for training purposes?",
        "What level of input and involvement do you expect from our deep learning team throughout the project?",
        "Can you provide a rough estimate of the scale of data you'll be working with (e.g., number of records, data volume)?",
        "Are there any intellectual property considerations or ownership rights related to the AI models or solutions we develop?",
        "What is your preferred frequency for project updates and reporting?",
        "Is there any other information or context you think would be valuable for us to know about your deep learning project?"
    ]

    
    def __init__(self):
        self.alienbabble={
            'describe_intent': r'Your planet.',
            'answer_why_intent': r'Why\sare.',
            'about_deep_learning': r'deep learning '
        }
    def greet(self):
        self.name=input('what is your name?\n')
        will_help=input(f'Hi {self.name}. I am Rule Bot. will you help me learn about your plan?\n')
        if will_help in self.negative_response:
            print('Ok. Have a nice earth day!')
            return 
        self.chat()
    
    def make_exit(self,reply):
        for command in self.exit_commands:
            if command==reply:
                print('Ok. Have A nice Day!')
                return True
    
    def chat(self):
        reply=input(random.choice(self.random_question)).lower()
        while not self.make_exit(reply):
            reply=input(self.match_reply(reply))
            
            
    def match_reply(self, reply):
        for key, value in self.alienbabble.items():
            intent=key
            regex_pattern=value
            found_match=re.match(regex_pattern,reply)
            if found_match and intent =='describe_intent':
                return self.describe_intent()
            elif found_match and intent =='answer_why_intent':
                return self.answer_why_intent()
            elif found_match and intent =='about_deep_learning':
                return self.about_deep_learning()
        if not found_match:
            return self.no_match_intent()
        
        
    def describe_intent(self):
        response=('I am a Data Scientist.\n','I am interested in Deep learning.\n','Now I am working on Computer Vision.\n','I also like NLP.\n ').lower()
        return random.choice(response)

    def answer_why_intent(self):
        response=('Data Science is a interesting sector.\n','I like to work on research.\n','Now a days Its become popular.\n','It\'s very easy and interesting to learn and explore .\n ').lower()
        return random.choice(response)

    def about_deep_learning(self):
        response=('Computer vision is about teaching computers to "see" and interpret visual information from the world, such as images and videos.\n ', 'Deep learning is a subset of machine learning using neural networks to automatically learn and make presiction from data.\n', 'NLP: Teaching computers to understand and use human language.\n', 'Machine learning: Computers learn from data to make decisions.\n', 'Artificiall inteligence computer mimicking human intelligence,\n','Data science: Extracting insights from data.\n')
        return random.choice(response)

    def no_match_intent(self):
        response=('Tell me more.\n','Please tell me more.\n','why do you say that?\n','I see. Can you elaborate?\n','Interesting . can you tell me more?\n','I see. how do you think?\n','Why?\n','How do you think I feel when you say that?\n')
        return random.choice(response)
```
